```python
from os import listdir
from matplotlib import image
from PIL import Image
import cv2
import os
```

```python
for filename in listdir('/content/anomalib/datasets/MVTec/firstiteration - Copy/test/annom/'):
  img_data = cv2.imread('/content/anomalib/datasets/MVTec/firstiteration - Copy/test/annom/'+ filename)
  image_resize = cv2.resize(img_data,(900,900))
  cv2.imwrite('/content/anomalib/datasets/MVTec/firstiteration - Copy/test/annom/'+ filename[:-3]+'png', image_resize)
  imagepath='/content/anomalib/datasets/MVTec/firstiteration - Copy/test/annom/'+filename
  os.remove(imagepath)

```
